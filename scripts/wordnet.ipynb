```python
from nltk.corpus import wordnet
from nltk.corpus import wordnet as wn
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from rdflib import Graph
import _pickle as cPickle
predicate_graph = cPickle.load(open('./wd_pred_prox_graph.pickle', 'rb'))

ent_type_vocab = dict()
data_predicate = list()
domain_vocab = dict()
ent_type_vocab_wordnet = dict()
range_vocab = dict()
for prox_t in  predicate_graph:
    prox_s,p,prox_o = prox_t
    for s in prox_s:
        if s not in ent_type_vocab:
            ent_type_vocab[s] = len(ent_type_vocab)
            try:
              ent_type_vocab_wordnet[s] = max([len(hyp_path) for hyp_path in wn.synsets(s)[0].hypernym_paths()])
            except:
              ent_type_vocab_wordnet[s] = 20
        
    for o in prox_o:
        if o not in ent_type_vocab:
            ent_type_vocab[o] = len(ent_type_vocab)
            try:
              ent_type_vocab_wordnet[o] = max([len(hyp_path) for hyp_path in wn.synsets(o)[0].hypernym_paths()])
            except:
              ent_type_vocab_wordnet[o] = 20


    
```

```python
import json
json_object = json.dumps(ent_type_vocab_wordnet, indent=4)

with open("./ent_type_vocab_wordnet.json", "w") as outfile:
  outfile.write(json_object)
```
